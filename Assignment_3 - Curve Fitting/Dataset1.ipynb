```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
csvfile = open('a3-data\dataset1.txt', 'r')
csvreader = csv.reader(csvfile, delimiter = ' ')

x = []
y = []

for row in csvreader:
    x.append(row[0])
    y.append(row[1])
```

```python
A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond = None)[0]
# Create a scatter plot of the data points
plt.scatter(x, y, label='Noisy Data', color = "blue")

# Create the regression line
regression_line = m * x + c
plt.plot(x, regression_line, color='red', label='Estimated')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Dataset1")
error_values = abs(regression_line - y)
plt.errorbar(x[::25], y[::25], yerr=error_values[::25], fmt='o', label='Error Bars', color='red')
plt.legend()

# Show the plot
plt.savefig("Dataset1.png")
plt.show()
```
